```python
#Loading libraries

import pandas as pd
import numpy as np
import pylab as pl
import matplotlib.pyplot as plt
#to suppresss the orange warnings
import warnings
warnings.filterwarnings("ignore")
```

__a. Load in the image data matrix (with rows as images and columns as features). Also load in the numeric class labels from the segmentation class file. Using your favorite method (e.g., sklearn's min-max scaler), perform min-max normalization on the data matrix so that each feature is scaled to [0,1] range.__

```python
#Read the data into panda's dataframe
segmentation_data = pd.read_csv("segmentation_data/segmentation_data.txt", header=None)
segmentation_data.head(5)
```

```python
segmentation_names = pd.read_csv("segmentation_data/segmentation_names.txt", header=None)
segmentation_names.head(5)
```

```python
segmentation_classes = pd.read_csv("segmentation_data/segmentation_classes.txt", header=None, sep='\t')
segmentation_classes.head()
```

```python
segmentation_data.columns
```

```python
from sklearn import preprocessing
```

```python
min_max_scaler = preprocessing.MinMaxScaler().fit(segmentation_data)
min_max_norm = min_max_scaler.transform(segmentation_data)
min_max_norm.shape
```

__b. Next, Perform Kmeans clustering (for this problem, use the Kmeans implementation in scikit-learn) on the image data (since there are a total 7 pre-assigned image classes, you should use K = 7 in your clustering). Use Euclidean distance as your distance measure for the clustering. Print the cluster centroids (use some formatting so that they are visually understandable). Compare your 7 clusters to the 7 pre-assigned classes by computing the Completeness and Homogeneity values of the generated clusters.__

```python
from sklearn.cluster import KMeans 
```

```python
kmeans = KMeans(n_clusters=7, max_iter=500, verbose=1) # initialization
```

```python
kmeans.fit(min_max_norm)
```

```python
clusters = kmeans.predict(min_max_norm)
```

```python
print (clusters)
```

```python
print (kmeans.cluster_centers_)
```

```python
#Cluster 0-6 represented with names
pd.DataFrame(kmeans.cluster_centers_.T, index=segmentation_names)
```

```python
from sklearn.metrics import completeness_score, homogeneity_score
```

```python
print (completeness_score(segmentation_classes[1],clusters))
```

```python
print (homogeneity_score(segmentation_classes[1],clusters))
```

__c. Perform PCA on the normalized image data matrix. You may use the linear algebra package in Numpy or the Decomposition module in scikit-learn (the latter is much more efficient). Analyze the principal components to determine the number, r, of PCs needed to capture at least 95% of variance in the data. Then use these r components as features to transform the data into a reduced dimension space.__

```python
from sklearn import decomposition
```

```python
pca = decomposition.PCA(n_components=5)
SegTrans = pca.fit(min_max_norm).transform(min_max_norm)
```

```python
np.set_printoptions(precision=2,suppress=True)
print (SegTrans)
```

```python
print(pca.explained_variance_ratio_)
```

*above makes it 98% and we will reduce 1 component to get to 95%*

```python
pca = decomposition.PCA(n_components=6)
SegTrans = pca.fit(min_max_norm).transform(min_max_norm)
```

```python
print(pca.explained_variance_ratio_.sum()*100)
```

```python
SegTrans.shape
```

*We went from 19 to 4 attributes space*

__d. Perform Kmeans again, but this time on the lower dimensional transformed data. Then, compute the Completeness and Homogeneity values of the new clusters.__

```python
kmeans.fit(SegTrans)
```

```python
clusters = kmeans.predict(SegTrans)
```

```python
print (kmeans.cluster_centers_)
```

```python
print (completeness_score(segmentation_classes[1],clusters))
```

```python
print (homogeneity_score(segmentation_classes[1],clusters))
```
